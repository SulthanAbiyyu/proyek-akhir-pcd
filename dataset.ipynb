```python
import os
import glob

import matplotlib.pyplot as plt
from pathlib import Path
from PIL import Image
```

```python
ROOT_DATASET_PATH = Path('./cifake/')
TRAIN_DATASET_PATH = ROOT_DATASET_PATH / 'train'
TEST_DATASET_PATH = ROOT_DATASET_PATH / 'test'
```

```python
train_fake = glob.glob(str(TRAIN_DATASET_PATH / 'FAKE' / '*.jpg'))
train_real = glob.glob(str(TRAIN_DATASET_PATH / 'REAL' / '*.jpg'))

print(f"Number of fake images in train dataset: {len(train_fake)}")
print(f"Number of real images in train dataset: {len(train_real)}")
```

```python
test_fake = glob.glob(str(TEST_DATASET_PATH / 'FAKE' / '*.jpg'))
test_real = glob.glob(str(TEST_DATASET_PATH / 'REAL' / '*.jpg'))

print(f"Number of fake images in test dataset: {len(test_fake)}")
print(f"Number of real images in test dataset: {len(test_real)}")
```

```python
NUMBER_OF_SAMPLES = 5

# fig axes NUMBER_OF_SAMPLES (column) x 2 (row)

fig, axes = plt.subplots(2, NUMBER_OF_SAMPLES, figsize=(10, 5))
axes[0, 2].set_title('Fake', fontweight="bold", size=15, pad=10)
axes[1, 2].set_title('Real', fontweight="bold", size=15,pad=10)

for i in range(NUMBER_OF_SAMPLES):
    axes[0, i].imshow(Image.open(train_fake[i]))
    axes[0, i].axis('off')
    axes[1, i].imshow(Image.open(train_real[i]))
    axes[1, i].axis('off')
```
